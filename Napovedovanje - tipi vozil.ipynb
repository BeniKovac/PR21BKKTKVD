```python
data = pd.concat([data_2000, data_2001, data_2002, data_2003, data_2004, data_2005, data_2006, data_2007, data_2008, data_2009, data_2010, data_2011, data_2012, 
                  data_2013, data_2014, data_2015, data_2016, data_2017, data_2018, data_2019], ignore_index = True)
data = data.fillna(0)
```

Z uporabo različnih metod bomo poskusili napovedati količino vsakega tipa vozil. Uporabili bomo linearno regresijo.

```python
mse_linear = []
rmse_linear = []
mae_linear = []

for tip_vozila in data:
    temp_mse = [] # za shranjevanje mse iz posamezne iteracije
    temp_mae = []
    temp_rmse = []
    for i in range(5): # 5x izracunaj
        temp_hx = [] # za shranjevanje predikcij
        X = data.loc[:, data.columns != tip_vozila]
        Y = data[tip_vozila]
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
        reg = LinearRegression()
        reg.fit(X_train, Y_train)
        for i in range(len(X_test.index)): 
            hx = reg.predict([X_test.values.tolist()[i]])
            temp_hx.append(hx[0])
        mse = mean_squared_error(temp_hx, Y_test)
        rmse = sqrt(mse)
        mae = mean_absolute_error(temp_hx, Y_test)
        temp_mse.append(mse)
        temp_rmse.append(rmse)
        temp_mae.append(mae)
    
    # dodaj povprecje teh petih poskusov
    mse_linear.append(np.mean(temp_mse))
    rmse_linear.append(np.mean(temp_rmse))
    mae_linear.append(np.mean(temp_mae))
```

```python
np.mean(mse_linear), np.mean(rmse_linear), np.mean(mae_linear)
```
