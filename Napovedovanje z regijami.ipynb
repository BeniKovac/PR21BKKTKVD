```python
from scipy.stats import pearsonr
def plot_fit_residual(x, y, yp):

    # Model
    fig, axes  = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
    axes[0].plot(x.ravel(), y.ravel(), "k.",  label="Podatki")
    axes[0].plot(x.ravel(), yp.ravel(), "g-", label="Model h(x)")
    axes[0].set_xlabel("x")
    axes[0].set_ylabel("y")
    #axes[0].set_xticks(data.values())
    #axes[0].set_xticklabels(labels, rotation=90)
    axes[0].legend(loc=4)

    # Ostanki
    r = pearsonr(y.ravel(), y.ravel()-yp.ravel())[0]
    axes[1].plot(y.ravel(), y.ravel()-yp.ravel(), "k.", label="Ostanek")
    axes[1].set_xlabel("y")
    axes[1].set_ylabel("y-h(x)")
    axes[1].set_title("Graf ostankov, R=%.3f" % r)
    axes[1].legend(loc=4)
    plt.show()
```

```python
x = np.fromiter(data.keys(), dtype=float)
y = np.fromiter(data.values(), dtype=float)
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
# Ucenje modela
model = LinearRegression()

model.fit(x, y)



# Napoved vrednosti za podatke
hx = model.predict(x)

plot_fit_residual(x, y, hx)
```

```python
i = mean_squared_error(hx, y)
i
```

```python
X = np.zeros((len(x), 2))
X[:, 0] = x.ravel()
X[:, 1] = x.ravel() ** 2 
# Učenje
model = LinearRegression()
model.fit(X, y)

# Napoved
hx = model.predict(X)

plot_fit_residual(x, y, hx)
```
