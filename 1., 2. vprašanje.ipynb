**Ali se je od leta 2000 do leta 2019 obremenjenost državnega cestnega omrežja Slovenije kaj spremenila?**

```python
obremenjenost = dict()
```

```python
obremenjenost[2000] = data_2000.sum()["Vsa vozila (PLDP)"]
obremenjenost[2001] = data_2001.sum()["Vsa vozila (PLDP)"]
obremenjenost[2002] = data_2002.sum()["Vsa vozila (PLDP)"]
obremenjenost[2003] = data_2003.sum()["Vsa vozila (PLDP)"]
obremenjenost[2004] = data_2004.sum()["Vsa vozila (PLDP)"]
obremenjenost[2005] = data_2005.sum()["Vsa vozila (PLDP)"]
obremenjenost[2006] = data_2006.sum()["Vsa vozila (PLDP)"]
obremenjenost[2007] = data_2007.sum()["Vsa vozila (PLDP)"]
obremenjenost[2008] = data_2008.sum()["Vsa vozila (PLDP)"]
obremenjenost[2009] = data_2009.sum()["Vsa vozila (PLDP)"]
obremenjenost[2010] = data_2010.sum()["Vsa vozila (PLDP)"]
obremenjenost[2011] = data_2011.sum()["Vsa vozila (PLDP)"]
obremenjenost[2012] = data_2012.sum()["Vsa vozila (PLDP)"]
obremenjenost[2013] = data_2013.sum()["Vsa vozila (PLDP)"]
obremenjenost[2014] = data_2014.sum()["Vsa vozila (PLDP)"]
obremenjenost[2015] = data_2015.sum()["Vsa vozila (PLDP)"]
obremenjenost[2016] = data_2016.sum()["Vsa vozila (PLDP)"]
obremenjenost[2017] = data_2017.sum()["Vsa vozila (PLDP)"]
obremenjenost[2018] = data_2018.sum()["Vsa vozila (PLDP)"]
obremenjenost[2019] = data_2019.sum()["Vsa vozila (PLDP)"]
```

```python
obremenjenost
```

```python
leta = obremenjenost.keys()
vrednosti = obremenjenost.values()
```

```python
plt.figure(figsize=(20,11))
plt.plot(leta, vrednosti, color="orange")
plt.xlabel("Leta")
plt.ylabel("Obremenjenost")
plt.xticks(range(2000,2020), rotation=60)
plt.title("Obremenjenost po letih, 2000 do 2019")
```

Iz grafa je vidno, da je obremenjenost od leta 2000 do 2010 skoraj neprestano naraščala. Med letoma 2011 in 2012 je padala, nato pa je začela spet naraščati. Od leta 2012 obremenjenost počasneje narašča. Domnevamo, da zato, ker so se ljudje začeli bolj osveščati o okolju in poskušajo najti druge, alternativne možnosti prevoza, ki zmanjšujejo obremenjenost.
