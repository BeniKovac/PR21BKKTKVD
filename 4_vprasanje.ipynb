```python
headers = set()
merilna_mesta = set()
merilna_mesta_na_zacetku = set()
popolna_merilna_mesta = set()

for i in range(2000, 2020, 1):
    with open(path+str(i)+'.csv', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=';', quotechar='"')
        # na zacetku vsakega CSVja je BoM, zato v naslednjem koraku uporabimo [3:]
        vrstica = next(reader)

        # print(', '.join(vrstica)[3:])
        if len(headers) > 0:
            headers = headers.intersection(set(vrstica))
        else:
            headers = set(vrstica)

        trenutna_merilna_mesta = set()
        for vrstica in reader:
            if i == 2000:
                merilna_mesta_na_zacetku.add(vrstica[6])
            merilna_mesta.add(vrstica[6])
            trenutna_merilna_mesta.add(vrstica[6])

        if i == 2000:
            popolna_merilna_mesta = merilna_mesta_na_zacetku
        popolna_merilna_mesta = popolna_merilna_mesta.intersection(trenutna_merilna_mesta)

print(popolna_merilna_mesta)
print(len(popolna_merilna_mesta))
popolna_merilna_mesta = sorted(list(popolna_merilna_mesta))
```

```python
for i in range(2000, 2020):
    trenutni = dfs[i]
    for index, row in main.iterrows():
        #if trenutni.index.dtype == 'float64':
        temp = trenutni.loc[int(index)]
        #else:
            #temp = trenutni.loc[index]
        #print(temp)
        if type(temp['Vsa vozila (PLDP)']) != np.int64:
            #print(temp)
            main.at[index, i] = temp['Vsa vozila (PLDP)'].iloc[0]
        else:
            main.at[index, i] = temp['Vsa vozila (PLDP)']
```

```python
spremembe = pd.DataFrame(columns= [a for a in range(2001,2020)])
```

```python
for i in range(2001, 2020):
    for index, row in main.iterrows():
        spremembe.at[index, i] = (row[i]-row[i-1])/(row[i]/100)
```

```python
spremembe.mean()
```

```python
plt.figure(figsize=(20,11))
plt.plot(spremembe.mean())
plt.xlabel("Leta")
plt.ylabel("Povprečna sprememba v %")
plt.xticks(range(2001,2021), rotation=60)
plt.title("Povprečna letna sprememba obremenjenosti")
```
