## Ugotavljanje kateri stolpci so skupni vsem dataframe-om.

```python
d = [list(data_2000.columns),list(data_2001.columns),list(data_2002.columns),list(data_2003.columns),
     list(data_2004.columns),list(data_2005.columns),list(data_2006.columns),list(data_2007.columns),
     list(data_2008.columns),list(data_2009.columns),list(data_2010.columns),list(data_2011.columns),
     list(data_2012.columns),list(data_2013.columns),list(data_2014.columns),list(data_2015.columns),
     list(data_2016.columns),list(data_2017.columns),list(data_2018.columns),list(data_2019.columns)]

skupni_atributi = set.intersection(*[set(x) for x in d])  
skupni_atributi
```

```python
s = []
s.append(data_2000)
s.append(data_2001)
s.append(data_2002)
s.append(data_2003)
s.append(data_2004)
s.append(data_2005)
s.append(data_2006)
s.append(data_2007)
s.append(data_2008)
s.append(data_2009)
s.append(data_2010)
s.append(data_2011)
s.append(data_2012)
s.append(data_2013)
s.append(data_2014)
s.append(data_2015)
s.append(data_2016)
s.append(data_2017)
s.append(data_2018)
s.append(data_2019)
for df in s:
    for column in df.columns:
        if column not in skupni_atributi:
            df.drop(column, inplace=True, axis=1)
```

```python
result = pd.concat(s)
result['Avtobusi'] = result['Avtobusi'].fillna(0)
result['Motorji'] = result['Motorji'].fillna(0)
result.sort_values(by='Osebna vozila', ascending=False)
```

```python
groupedResult = result.groupby(['Prometni odsek']).agg('sum')
groupedResult.drop("Stev. odseka", inplace=True, axis=1)
groupedResult.drop("Stac. zacetka", inplace=True, axis=1)
groupedResult.sort_values(by='Tov. s prik.', ascending=False)
```

```python
list = ["Prometni odsek"]
list.append("Vsa vozila (PLDP)")
list.append("Motorji")
list.append("Osebna vozila")
list.append("Avtobusi")
list.append("Tez. tov. nad 7t")
list.append("Tov. s prik.")

data = []
for ime_odseka, row in groupedResult.iterrows():
    rating_list = []
    rating_list.append(ime_odseka)
    for i in range(6):
        rating_list.append(row[i])
    data.append(rating_list)

df = pd.DataFrame(data, columns=list)
df
```

##Dodajanje stolpca z vsoto vozil:

```python
import scipy.cluster.hierarchy as sch
import scipy
#L = sch.linkage(dfDropped, method="average", metric="cosine")
dfDropped = df.drop('Prometni odsek', axis=1)
dfDropped = dfDropped.drop(dfDropped.index[456])
a_series = (dfDropped == 0).all(axis=1)
new_df = dfDropped.loc[a_series]
new_df
```

```python
sumlist = []
for index, row in dfDropped.iterrows():
    sumlist.append(sum(row))
sumlist
dfDropped['SumOfRow'] = sumlist
dfDropped
```

#Dendrogram 100 najbolj prevoženih odsekov:

Z uporabo kosinusne razdalje za mero in povprečne razdalje za metodo:

```python
dfDropped100 = dfDropped.sort_values(by='SumOfRow').head(100)
L = sch.linkage(dfDropped100, method="average", metric="cosine")

plt.figure(figsize=(25, 16))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
plt.show()
```

Z uporabo evklidske razdalje za mero in povprečne razdalje za metodo:

```python
L = sch.linkage(dfDropped100, method="average", metric="euclidean")

plt.figure(figsize=(25, 16))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
plt.show()
```

Z uporabo Manhattanske razdalje za mero in povprečne razdalje za metodo:

```python
L = sch.linkage(dfDropped100, method="average", metric="cityblock")

plt.figure(figsize=(25, 16))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
plt.show()
```

Z uporabo Evklidske razdalje za mero in razdalje med središčema za metodo:

```python
L = sch.linkage(dfDropped100, method="centroid", metric="euclidean")

plt.figure(figsize=(25, 16))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
plt.show()
```

Z uporabo Evklidske razdalje za mero in razdalje med najbližjima točkama gruč za metodo:

```python
L = sch.linkage(dfDropped100, method="single", metric="euclidean")

plt.figure(figsize=(25, 16))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
plt.show()
```

Z uporabo Manhattanske razdalje za mero in razdalje med najbližjima točkama gruč za metodo:

```python
L = sch.linkage(dfDropped100, method="single", metric="cityblock")

plt.figure(figsize=(25, 16))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
plt.show()
```

Z uporabo kosinusne razdalje za mero in razdalje med najbližjima točkama gruč za metodo:

```python
L = sch.linkage(dfDropped100, method="single", metric="cosine")

plt.figure(figsize=(25, 16))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
plt.show()
```

Najbolj zanimiv se nam zdi dendrogram z uporabo evklidske razdalje za mero in povprečne razdalje za metodo. Tega bomo torej uporabili za določitev gruč:

```python
L = sch.linkage(dfDropped100, method="average", metric="euclidean")

plt.figure(figsize=(25, 25))
labels = [df[df['Prometni odsek'] == x].values[0][0] for x in df['Prometni odsek'].values]
D      = sch.dendrogram(L, labels=labels, leaf_font_size=15)
plt.ylabel("Razdalja")
t = 48
plt.plot([0, 1000], [t, t], "k--")
plt.show()
```
